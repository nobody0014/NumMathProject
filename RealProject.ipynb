```python
%matplotlib inline
import scipy
import scipy.io.wavfile
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import Audio
```

```python
rate, data = scipy.io.wavfile.read("athing.wav")
mono = data[:,0]
```

```python
#code copied for aj
def suppress(a,begin,last,factor):
    a_fft = np.fft.rfft(a)
    new_fft = a_fft.copy()
    new_fft[begin:last] = 0
    return np.fft.irfft(new_fft)
def double_chunk(mono):
    ret = np.zeros(len(mono),dtype = mono.dtype)
    chunksize = 20000
    for i in range(0,len(mono),chunksize):
        first = i
        chunk = mono[i:i+chunksize]
        last = i + len(chunk)
        ret[first:last] = suppress(chunk,0,50,100.0)
    return ret
```
